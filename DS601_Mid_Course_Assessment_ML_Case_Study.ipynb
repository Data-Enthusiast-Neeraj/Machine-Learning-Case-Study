# **Retail Sales Prediction: A Machine Learning Case Study**

##üìù **Project Overview**

**Objective**:

This project aims to forecast daily sales for Rossmann, a prominent retail pharmacy chain with over 3,000 stores across seven European countries. Accurate short-term sales predictions are vital for operational planning and promotional strategy. Store-level sales are influenced by various dynamic factors such as promotional campaigns, competitor proximity, school and public holidays, seasonal trends, and regional demographics. By leveraging historical data and external factors, we develop predictive models to support store managers in generating reliable six-week sales forecasts.
**Tools & Technologies:**

* Python (Pandas, NumPy, Matplotlib, Seaborn, Scikit-learn)

* Google Colab

* Time series analysis and machine learning techniques


**Outcome:**

Develop and evaluate machine learning models capable of accurately forecasting store-level sales over a six-week horizon. These predictions can assist Rossmann store managers in inventory management, staffing decisions, and promotional planning‚Äîultimately leading to more efficient operations and improved customer satisfaction

##‚ùó **Problem Statement**

We have access to historical sales data from 1,115 Rossmann stores. Our objective is to predict the "Sales" column for the test set. It's worth noting that some stores in the dataset were temporarily closed due to renovation work.

## üìä **Dataset Description**








The dataset includes two main files:

***Rossmann Stores Data.csv:*** This file contains historical data, including sales information.

***store.csv:*** This file provides supplemental details about the stores.
Data Fields:

***Id:*** An identifier representing a combination of store and date within the test set.
***Store:*** A unique identifier for each store.

***Sales:*** The turnover for a particular day (this is the target variable for prediction).

***Customers:*** The number of customers on a given day.

***Open:*** An indicator of whether the store was open (0 = closed, 1 = open).

***StateHoliday:*** Indicates a state holiday; typically, stores are closed on these days. Types include a = public holiday, b = Easter holiday, c = Christmas, 0 = None.

***SchoolHoliday:*** Indicates if the store was affected by the closure of public schools on that date.

***StoreType:*** Differentiates between four store models: a, b, c, d.

***Assortment:*** Describes the assortment level: a = basic, b = extra, c = extended.

***CompetitionDistance:*** The distance in meters to the nearest competitor store.

***CompetitionOpenSince[Month/Year]:*** Approximate year and month when the nearest competitor store opened.

***Promo:*** Indicates whether a store is running a promotion on that day.

***Promo2:*** Indicates whether the store is participating in a continuing and consecutive promotion (0 = not participating, 1 = participating).

***Promo2Since[Year/Week]:*** Describes the year and calendar week when the store started participating in Promo2.

***PromoInterval:*** Describes the consecutive intervals when Promo2 is initiated, listing the months the promotion begins anew. For example, "Feb,May,Aug,Nov" indicates that the promotion restarts in February, May, August, and November of any given year for that store.


## üîç **Methodology**



The project workflow follows these major steps:

1. **Data Collection & Understanding**

   * Utilized historical sales data provided by Rossmann, including store-specific attributes, promotional details, holiday schedules, and competition metrics.

2. **Exploratory Data Analysis (EDA)**

   * Conducted EDA to explore sales trends, seasonal patterns, and correlations between variables affecting store performance.

3. **Data Preprocessing**

   * Managed missing or inconsistent entries, encoded categorical variables such as store type and state holidays, and ensured date formats were consistent for time-series modeling.

4. **Feature Engineering**

   * Created new features including days since last promotion, competition duration, holiday indicators, and time-based components like day of the week and month.

5. **Model Selection & Training**

   * Applied machine learning models including Random Forest, Gradient Boosting (XGBoost), and Linear Regression. Employed cross-validation and grid search for optimal parameter tuning.

6. **Model Evaluation**

   * Evaluated model accuracy using RMSE and R¬≤ metrics, and compared performance across stores and time periods to ensure robustness.


7. **Conclusion and Future Work**

   * Generated six-week daily sales forecasts for each store, enabling Rossmann managers to make informed planning and inventory decisions.

# Removing Outliers Of Sales Column

```python
#removing outliers
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
```

```python
# defining new variable after removing outliers
df= remove_outlier(df, 'Sales')
```

```python
def rmse(x, y):
    return sqrt(mean_squared_error(x, y))

# definte MAPE function
def mape(x, y):
    return np.mean(np.abs((x - y) / x)) * 100

train_score_2=lasreg.score(X_train, y_train)
test_score_2=lasreg.score(X_test, y_test)

print("Regresion Model Score" , ":" , train_score_2 , "," ,
      "Out of Sample Test Score" ,":" , test_score_2)

y_predicted = lasreg.predict(X_train)
y_test_predicted = lasreg.predict(X_test)

print("Training RMSE", ":", rmse(y_train, y_predicted),
      "Testing RMSE", ":", rmse(y_test, y_test_predicted))
print("Training MAPE", ":", mape(y_train, y_predicted),
      "Testing MAPE", ":", mape(y_test, y_test_predicted))
```
